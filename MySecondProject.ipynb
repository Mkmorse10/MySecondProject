```python
import talib as ta 
import matplotlib.pyplot as plt
import yfinance as yf
from pandas_datareader import data
```

```python
start_date= "2019-01-01"
end_date="2020-01-01"
```

```python
amd=data.DataReader('AMD','yahoo',start_date,end_date)
```

```python
data=yf.download("AMD",start="2017-01-01",end="2020-07-23")
```

```python
data['Simple MA'] = ta.SMA(data['Close'], 14)
data['EMA']=ta.EMA(data['Close'],timeperiod = 14)
data[['Close','Simple MA','EMA']].plot(figsize=(15,15) )
plt.show()
```
